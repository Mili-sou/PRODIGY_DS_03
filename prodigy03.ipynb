```python
df.hist(figsize=(10,10),color='purple')
plt.show()
```

```python
# Calculate the number of rows and columns for subplots
numerical_plots = len(category_cols)
# Add 1 and divide by 2 to round up for odd numbers
numerical_rows = (numerical_plots + 1) // 2  
numerical_cols = 2

# Create a new figure
plt.figure(figsize=(20, 25)) 

# Loop through each feature and create a countplot
for i, feature in enumerate(category_cols, 1):
    plt.subplot(numerical_rows, numerical_cols, i)
    sns.countplot(x=feature, data=df, palette='rocket')
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)

# Adjust layout to prevent overlap of subplots
plt.tight_layout()
plt.show()
```

```python
column = df[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
df.plot(kind='box', subplots=True, layout=(2,5),figsize=(20,10),color='magenta')
plt.show()
```

```python
# Exclude non-numeric columns
numeric_df = df.drop(columns=category_cols)

# Compute the correlation matrix
corr = numeric_df.corr()

# Print the correlation matrix
print(corr)

# Filter correlations with absolute value >= 0.90
corr = corr[abs(corr) >= 0.90]

sns.heatmap(corr,annot=True,cmap='coolwarm',linewidths=0.2)
plt.show()
```

```python
high_corr_cols = ['emp.var.rate','euribor3m','nr.employed']
df1 = df.copy()
# axis=1 indicates columns
df1.drop(high_corr_cols,inplace=True,axis=1)  
df1.columns
```

# Label Encoding

```python
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['deposit'].value_counts()
```

```python
# independent variable
x = df_encoded.drop('deposit',axis=1)  
# dependent variable
y = df_encoded['deposit']             
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

# Training and testing of dataset 

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  
    print('Testing Score',test_score)

dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```

# Decision tree

```python
cl = ['no','yes']
fn = x_train.columns
print(fn)
print(cl)
```

```python
plt.figure(figsize=(30,10))
plot_tree(dt,class_names=cl,filled=True)
plt.show()
```

# Decision Tree(on Entropy criteria)

```python
dt1 = DecisionTreeClassifier(criterion='entropy',max_depth=4,min_samples_split=15)
dt1.fit(x_train,y_train)
```

```python
mscore(dt1)
```

```python
ypred_dt1 = dt1.predict(x_test)
eval_model(y_test,ypred_dt1)
```
